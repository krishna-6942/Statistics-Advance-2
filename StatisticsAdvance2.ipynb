**THEORY QUESTIONS**

1. What is hypothesis testing in statistics?
- Hypothesis testing is a method in statistics used to make decisions or inferences about population parameters based on sample data. It tests assumptions (hypotheses) to determine if observed effects are statistically
significant.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
- The null hypothesis (H₀) states there is no effect or difference, while the alternative hypothesis (H₁) suggests there is an effect. Hypothesis testing seeks evidence to reject H₀ in favor of H₁.

3. What is the significance level in hypothesis testing, and why is it important?
- The significance level (α) is the threshold for rejecting H₀, commonly set at 0.05. It represents the probability of making a Type I error (false positive).


4. What does a P-value represent in hypothesis testing?
- A P-value is the probability of obtaining results as extreme as the observed results, assuming H₀ is true. It helps determine the strength of evidence against H₀.


5. How do you interpret the P-value in hypothesis testing?
- A small P-value (≤ α) indicates strong evidence against H₀, leading to its rejection. A large P-value (> α) suggests weak evidence, so H₀ is not rejected.

6. What are Type 1 and Type 2 errors in hypothesis testing?
- Type 1 error is rejecting a true H₀ (false positive), while Type 2 error is failing to reject a false H₀ (false negative). Balancing both is important in test design.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
- A one-tailed test checks for an effect in one direction, while a two-tailed test checks in both directions. The choice depends on the research question.


8. What is the Z-test, and when is it used in hypothesis testing?
- A Z-test is used when the population standard deviation is known and the sample size is large (n > 30). It's ideal for testing means and proportions.


9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
- The Z-score is calculated as (X - μ)/σ and shows how many standard deviations an observation is from the mean. It's used in standard normal distribution tests.


10. What is the T-distribution, and when should it be used instead of the normal distribution?
- The T-distribution is used instead of the normal distribution when the sample size is small and the population standard deviation is unknown.


11. What is the difference between a Z-test and a T-test?
- A Z-test uses the normal distribution and requires known variance and a large sample. A T-test is used with small samples and unknown variance.


12. What is the T-test, and how is it used in hypothesis testing?
- A T-test compares sample means to determine if they differ significantly, often used in comparing two groups or before/after scenarios.


13. What is the relationship between Z-test and T-test in hypothesis testing?
- Both tests compare means, but the Z-test is for large samples/known variance, and the T-test for small samples/unknown variance.


14. What is a confidence interval, and how is it used to interpret statistical results?
- A confidence interval (CI) estimates a range within which a population parameter lies, with a given confidence level (e.g., 95%). It reflects uncertainty.


15. What is the margin of error, and how does it affect the confidence interval?
- The margin of error indicates the range of uncertainty in the estimate. Larger margins mean less precision and wider confidence intervals.


16. How is Bayes' Theorem used in statistics, and what is its significance?
- Bayes' Theorem updates probabilities based on new evidence. It is vital in inferential statistics and decision-making under uncertainty.


17. What is the Chi-square distribution, and when is it used?
- The Chi-square distribution is used in tests of independence and goodness-of-fit, especially for categorical data.


18. What is the Chi-square goodness of fit test, and how is it applied?
- The Chi-square goodness-of-fit test assesses if observed frequencies differ from expected ones, helping evaluate model fit.


19. What is the F-distribution, and when is it used in hypothesis testing?
- The F-distribution is used in comparing two variances, commonly in ANOVA and regression analysis.


20. What is an ANOVA test, and what are its assumptions?
- An ANOVA (Analysis of Variance) test checks whether there are statistically significant differences between means of three or more groups. It assumes normality, independence, and equal variances.


21. What are the different types of ANOVA tests?
- Types of ANOVA include one-way ANOVA (one factor), two-way ANOVA (two factors), and repeated measures ANOVA (same subjects tested repeatedly).


22. What is the F-test, and how does it relate to hypothesis testing?
- The F-test is used in ANOVA and regression to compare variances and test if the group means are significantly different.

1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results.

```python
from scipy import stats
import numpy as np

# Sample data
sample = [20, 22, 23, 19, 21, 20, 22]
pop_mean = 21
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
n = len(sample)

# Z-test calculation
z = (sample_mean - pop_mean) / (sample_std / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print(f"Z-value: {z:.3f}")
print(f"P-value: {p_value:.3f}")
```

3.  Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

```python
def one_sample_z_test(data, pop_mean):
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    n = len(data)
    z = (sample_mean - pop_mean) / (sample_std / np.sqrt(n))
    p = 2 * (1 - stats.norm.cdf(abs(z)))
    return z, p

sample = [10, 12, 9, 11, 13, 12]
z, p = one_sample_z_test(sample, 10)
print(f"Z-score: {z:.3f}, P-value: {p:.3f}")
```

4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import matplotlib.pyplot as plt

z = 1.96  # example value
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

plt.plot(x, y)
plt.fill_between(x, y, where=(x <= -z) | (x >= z), color='red', alpha=0.5)
plt.title('Two-tailed Z-test: Rejection Region')
plt.xlabel('Z')
plt.ylabel('Probability Density')
plt.axvline(-z, color='red', linestyle='--')
plt.axvline(z, color='red', linestyle='--')
plt.show()
```

5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
def plot_type_errors():
    x = np.linspace(-4, 8, 1000)
    null_dist = stats.norm(loc=0, scale=1)
    alt_dist = stats.norm(loc=2, scale=1)

    plt.plot(x, null_dist.pdf(x), label='Null Hypothesis (H₀)', color='blue')
    plt.plot(x, alt_dist.pdf(x), label='Alternative Hypothesis (H₁)', color='green')

    alpha = 0.05
    critical_value = stats.norm.ppf(1 - alpha)
    plt.axvline(critical_value, color='red', linestyle='--', label='Critical Value')

    plt.fill_between(x, null_dist.pdf(x), where=(x >= critical_value), color='blue', alpha=0.3, label='Type I Error')
    plt.fill_between(x, alt_dist.pdf(x), where=(x < critical_value), color='green', alpha=0.3, label='Type II Error')

    plt.legend()
    plt.title('Type I and Type II Errors')
    plt.xlabel('Test Statistic')
    plt.ylabel('Density')
    plt.grid(True)
    plt.show()

plot_type_errors()
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    sem = stats.sem(data)
    margin = sem * stats.t.ppf((1 + confidence) / 2., len(data)-1)
    lower = mean - margin
    upper = mean + margin
    return lower, upper

sample = [5.1, 5.3, 5.2, 5.5, 5.0, 5.4]
ci = confidence_interval(sample)
print(f"95% Confidence Interval: {ci}")
```

10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

```python
def margin_of_error(data, confidence=0.95):
    sem = stats.sem(data)
    moe = sem * stats.t.ppf((1 + confidence) / 2., len(data)-1)
    return moe

sample_data = [120, 122, 119, 121, 123, 118]
moe = margin_of_error(sample_data)
print(f"Margin of Error: {moe:.3f}")
```

11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
def bayes_theorem(prior_A, prob_B_given_A, prob_B_given_not_A):
    prob_not_A = 1 - prior_A
    prob_B = prob_B_given_A * prior_A + prob_B_given_not_A * prob_not_A
    posterior_A_given_B = (prob_B_given_A * prior_A) / prob_B
    return posterior_A_given_B

# Example: Medical test
prior = 0.01               # P(Disease)
sensitivity = 0.95         # P(Positive|Disease)
false_positive = 0.10      # P(Positive|No Disease)

posterior = bayes_theorem(prior, sensitivity, false_positive)
print(f"Posterior probability of having disease given a positive test: {posterior:.4f}")
```

13.  Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data.

```python
def expected_frequencies(observed):
    observed = np.array(observed)
    row_totals = observed.sum(axis=1)
    col_totals = observed.sum(axis=0)
    total = observed.sum()
    expected = np.outer(row_totals, col_totals) / total
    return expected

observed = [[30, 10], [20, 40]]
expected = expected_frequencies(observed)
print("Expected Frequencies:")
print(np.round(expected, 2))
```

16. Implement an F-test using Python to compare the variances of two random samples.

```python
def f_test(var1, var2, n1, n2):
    f = var1 / var2
    dfn, dfd = n1 - 1, n2 - 1
    p_value = 1 - stats.f.cdf(f, dfn, dfd)
    return f, p_value

sample1 = [10, 12, 11, 13, 12]
sample2 = [8, 9, 7, 10, 9]

f_stat, p = f_test(np.var(sample1, ddof=1), np.var(sample2, ddof=1), len(sample1), len(sample2))
print(f"F-statistic: {f_stat:.3f}, P-value: {p:.3f}")
```

19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

```python
import seaborn as sns
from scipy.stats import shapiro, levene

def check_anova_assumptions(groups):
    all_data = np.concatenate(groups)
    labels = np.concatenate([[f'G{i+1}']*len(g) for i, g in enumerate(groups)])
    df = pd.DataFrame({'Values': all_data, 'Group': labels})

    # Normality test (Shapiro)
    for i, g in enumerate(groups):
        stat, p = shapiro(g)
        print(f"Group {i+1} Normality P-value: {p:.3f}")

    # Equal variance (Levene's test)
    stat, p = levene(*groups)
    print(f"Levene’s test for equal variance P-value: {p:.3f}")

    sns.boxplot(x='Group', y='Values', data=df)
    plt.title("Boxplot for ANOVA Assumptions")
    plt.show()

check_anova_assumptions([group1, group2, group3])
```

24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

```python
sample = np.array([14, 16, 15, 14, 15, 16, 17])
n = len(sample)
sample_var = np.var(sample, ddof=1)
pop_var = 4  # hypothesized population variance

chi2_stat = (n - 1) * sample_var / pop_var
p_value = 1 - stats.chi2.cdf(chi2_stat, df=n-1)

print(f"Chi-square Statistic: {chi2_stat:.3f}, P-value: {p_value:.3f}")
```
